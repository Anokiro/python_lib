#### Тема “Визуализация данных в Matplotlib” Задание 1.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

x = [1, 2, 3, 4, 5, 6, 7 ]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]

plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y, label='Cписки x и y', linestyle='dotted', color='red', facecolor='grey')
plt.xlabel('this is x', alpha=0.2)
plt.ylabel('this is y', alpha=0.2)
plt.grid(alpha=0.2)
plt.legend(frameon=False)

title_dict = {'family': 'Arial', 'fontsize': 14, 'color': '#909900'}
plt.title('Диаграмма рассеяния', fontdict=title_dict)
```

#### Тема “Визуализация данных в Matplotlib” Задание 2.

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
plt.plot(t, f, color='green', linewidth=8)

plt.title('График f(t)')
plt.xlabel('Значения t', alpha=0.2)
plt.ylabel('Значения f', alpha=0.2)
plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

#### Тема “Визуализация данных в Matplotlib” Задание 3.

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
plt.subplots_adjust(wspace=0.3, hspace=0.3) 
fig.set_size_inches(8, 6)

ax1.set_title('График 1')
ax2.set_title('График 2')
ax3.set_title('График 3')
ax4.set_title('График 4')

ax1.set_xlim([-5, 5])
```

#### Тема “Визуализация данных в Matplotlib” Задание 4.

```python
import pandas as pd
myset = pd.read_csv('C:/Users/Anokiro/creditcard.csv', sep= ',')
myset.head(10)
```

```python
my_x = myset['Class'].value_counts()
plt.style.use('fivethirtyeight')
my_x
```

```python
my_x.plot(kind='bar', logy=True)
```

```python
v_zero = myset.loc[(myset['Class'] == 0), 'V1']
v_one = myset.loc[(myset['Class'] == 1), 'V1']

plt.hist(v_zero, bins = 20, label='Class(V1)=0', density=True, edgecolor='white', color='grey')

plt.hist(v_one, bins = 20, label='Class(V1) = 1', alpha=0.4, density=True, edgecolor='black', color='red')
plt.xlabel('Class', alpha=0.5)
plt.legend(fontsize=8, frameon=False)
```

```python
#Поэкспериментируем с другой библиотекой (за рамками задания 4)
import seaborn as sns
plt.figure(figsize=(5, 2))
sns.kdeplot(v_zero, shade=True, color="g", label="=0", alpha=.7)
sns.kdeplot(v_one, shade=True, color="r", label="=1", alpha=.5)

plt.legend()
```
